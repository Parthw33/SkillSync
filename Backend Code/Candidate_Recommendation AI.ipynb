```python
# Function to extract text from a PDF using pdfplumber
def extract_text_from_pdf(pdf_path):
    with pdfplumber.open(pdf_path) as pdf:
        text = ''
        for page in pdf.pages:
            text += page.extract_text()
    return text

# Upload PDF to Google Colab (you will be prompted to upload a file)
from google.colab import files
uploaded = files.upload()

# Extract the uploaded PDF file
pdf_path = list(uploaded.keys())[0]
pdf_text = extract_text_from_pdf(pdf_path)

print("Extracted Job Description Text:\n", pdf_text)
```

```python
# Define a list of possible skills
skill_keywords = [
    "python", "machine learning", "aws", "tensorflow", "java",
    "data analysis", "sql", "react", "javascript", "html", "css"
]

# Function to extract relevant skills from the job description text
def extract_skills_from_text(text, skill_keywords):
    skills_found = []
    text_lower = text.lower()  # Convert text to lowercase for matching
    for skill in skill_keywords:
        if skill in text_lower:
            skills_found.append(skill)
    return skills_found

# Extract skills from the job description text
job_skills_list = extract_skills_from_text(pdf_text, skill_keywords)

print("Extracted Job Skills:", job_skills_list)
```
